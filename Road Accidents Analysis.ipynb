### Q1. The percentage of road accidents during all the years.

```python
data= [df["State/UT-Wise Total Number of Road Accidents during - 2013"].sum(),
       df["State/UT-Wise Total Number of Road Accidents during - 2014"].sum(), 
       df["State/UT-Wise Total Number of Road Accidents during - 2015"].sum(), 
       df["State/UT-Wise Total Number of Road Accidents during - 2016"].sum()]

my_labels=["State/UT-Wise Total Number of Road Accidents during - 2013",
           "State/UT-Wise Total Number of Road Accidents during - 2014",
           "State/UT-Wise Total Number of Road Accidents during - 2015",
           "State/UT-Wise Total Number of Road Accidents during - 2016"]

plt.title('Percentage of road accidents during all the years',fontsize=15, color="blue")

plt.pie(data, labels=my_labels, autopct="%1.1f%%", colors=sns.color_palette("flare"))
plt.show()
```

Inference: From above presentation we conclude that 
           State/UT-Wise Total Number of Road Accidents during - 2013= 24.8% ,
           State/UT-Wise Total Number of Road Accidents during - 2014= 25%,
           State/UT-Wise Total Number of Road Accidents during - 2015= 25.6%, 
           State/UT-Wise Total Number of Road Accidents during - 2016= 24.5%

### Q2. Mean Accidents per 1L population for each year.

```python
mean_data= [df["Total Number of Accidents Per Lakh Population - 2013"].mean(),
       df["Total Number of Accidents Per Lakh Population - 2014"].mean(), 
       df["Total Number of Accidents Per Lakh Population - 2015"].mean(), 
       df["Total Number of Accidents Per Lakh Population - 2016"].mean()]

labels=["2013","2014","2015","2016"]

plt.title('Mean Accidents per 1L population for each year.',fontsize=15, color="black")

plt.pie(mean_data, labels=labels, autopct="%1.1f%%", colors=sns.color_palette("blend:#7AB,#EDA"))
plt.show()
```

Inference: From above presentation we conclude that Mean Accidents per 1L population for year 2013 is 25.8%, 2014 is 24.6%, 2015 is 25.2%, 2016 is 24.4%
